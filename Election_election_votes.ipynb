```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("election_separate_votes.csv")
df
```

```python
print(df.columns)
```

```python
parties = ["Votes_BJP", "Votes_INC", "Votes_RJD", "Votes_JDU", "Votes_LJP"]
```

```python
for party in parties:
    yearwise = df.groupby("Year")[party].sum()
    plt.plot(yearwise.index, yearwise.values, marker="o", label=party)

plt.title("Party Vote Trends (20202025)")
plt.xlabel("Year")
plt.ylabel("Total Votes")
plt.legend()
plt.grid(True)
plt.show()
```

```python
bjp_votes = df.groupby("Year")["Votes_BJP"].sum()
inc_votes = df.groupby("Year")["Votes_INC"].sum()

plt.bar(bjp_votes.index - 0.2, bjp_votes.values, width=0.4, label="BJP",)
plt.bar(inc_votes.index + 0.2, inc_votes.values, width=0.4, label="INC")

plt.title("BJP vs INC Votes Comparison")
plt.xlabel("Year")
plt.ylabel("Total Votes")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("election_separate_votes.csv")

print("First 5 rows of the dataset:")
print(df.head())
print()

# 1) Top parties winning seats

top_parties = df["Winning_Party"].value_counts().head(10)
print("Top parties winning seats:")
print(top_parties)
print()

# 2) Turnout comparison (2020 vs 2025)

turnout_by_year = df.groupby("Year")["Turnout_Percentage"].mean()
print("Average Turnout by Year:")
print(turnout_by_year)
print()

# Plot Turnout Comparison
turnout_by_year.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Average Voter Turnout by Year")
plt.xlabel("Year")
plt.ylabel("Turnout Percentage")
plt.ylim(0, 100)
plt.grid(axis='y')
plt.show()


# 3) Summary statistics of vote share by district

summary_stats_by_district = df.groupby("District")[["Votes_BJP", "Votes_INC", "Votes_RJD", "Votes_JDU", "Votes_LJP"]].mean()
print("Average votes by district:")
print(summary_stats_by_district)
print()

# Optional: you can calculate winning & runner-up vote shares
df['Winning_VoteShare_Percent'] = df[['Votes_BJP','Votes_INC','Votes_RJD','Votes_JDU','Votes_LJP']].max(axis=1) / df['Total_Voters'] * 100
df['RunnerUp_VoteShare_Percent'] = df[['Votes_BJP','Votes_INC','Votes_RJD','Votes_JDU','Votes_LJP']].apply(lambda row: sorted(row, reverse=True)[1], axis=1) / df['Total_Voters'] * 100

summary_vote_share = df.groupby("District")[["Winning_VoteShare_Percent", "RunnerUp_VoteShare_Percent"]].mean()
print("Summary vote share by district:")
print(summary_vote_share)
print()

# 4) Trend of elections over years (number of constituencies)

election_trend = df.groupby("Year").size()
plt.plot(election_trend.index, election_trend.values, marker='o', linestyle='-', color='green')
plt.title("Trend of Election Constituencies Count (2020 vs 2025)")
plt.xlabel("Year")
plt.ylabel("Number of Constituencies")
plt.grid(True)
plt.show()
```

```python
parties = ["Votes_BJP", "Votes_INC", "Votes_RJD", "Votes_JDU", "Votes_LJP"]
party_colors = ["blue", "orange", "green", "purple", "red"]

# Vote trend per year
yearly_votes = df.groupby("Year")[parties].sum()

#  LINE PLOT: Vote Trends
plt.figure(figsize=(8, 5))
for i, party in enumerate(parties):
    plt.plot(yearly_votes.index, yearly_votes[party], marker='o', label=party, color=party_colors[i])

plt.title("Vote Trends by Party (20202025)")
plt.xlabel("Year")
plt.ylabel("Votes")
plt.legend()
plt.grid(True)
plt.savefig("line_vote_trend.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# --- 2) BAR CHART: Party Vote Comparison ---
plt.figure(figsize=(8, 5))
x = range(len(yearly_votes.index))
width = 0.15

for i, party in enumerate(parties):
    plt.bar([p + i * width for p in x], yearly_votes[party], width=width, label=party)

plt.xticks([p + width for p in x], yearly_votes.index)
plt.title("Party Votes by Year (Bar Chart)")
plt.xlabel("Year")
plt.ylabel("Votes")
plt.legend()
plt.savefig("bar_yearly_votes.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# PIE CHART: Total Vote Share 
total_vote_share = yearly_votes.sum()

plt.figure(figsize=(6, 6))
plt.pie(total_vote_share.values, labels=total_vote_share.index, autopct="%1.1f%%", startangle=90)
plt.title("Overall Vote Share (20202025)")
plt.savefig("pie_vote_share.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# HISTOGRAM: Turnout Distribution 
plt.figure(figsize=(8, 5))
plt.hist(df["Turnout_Percentage"], bins=10, edgecolor="black")
plt.title("Turnout Percentage Distribution")
plt.xlabel("Turnout %")
plt.ylabel("Frequency")
plt.savefig("turnout_histogram.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Data
parties = ["Votes_BJP", "Votes_INC", "Votes_RJD", "Votes_JDU", "Votes_LJP"]
party_colors = ["blue", "orange", "green", "purple", "red"]

# Grouped data
yearly_votes = df.groupby("Year")[parties].sum()
total_vote_share = yearly_votes.sum()
turnout_values = df["Turnout_Percentage"]

# --- PLOT DASHBOARD ---
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Bihar Election Analysis (20202025)", fontsize=16, fontweight="bold")

# Line Plot  Vote Trends
for i, party in enumerate(parties):
    axes[0, 0].plot(yearly_votes.index, yearly_votes[party], marker='o', label=party)

axes[0, 0].set_title("Party Vote Trends")
axes[0, 0].set_xlabel("Year")
axes[0, 0].set_ylabel("Votes")
axes[0, 0].grid(True)
axes[0, 0].legend()

# Bar Chart  Party Votes by Year
width = 0.15
x = range(len(yearly_votes.index))

for i, party in enumerate(parties):
    axes[0, 1].bar([p + i * width for p in x], yearly_votes[party], width=width, label=party)

axes[0, 1].set_xticks([p + width for p in x])
axes[0, 1].set_xticklabels(yearly_votes.index)
axes[0, 1].set_title("Year-wise Party Votes")
axes[0, 1].set_xlabel("Year")
axes[0, 1].set_ylabel("Votes")
axes[0, 1].legend()

# Pie Chart  Vote Share
axes[1, 0].pie(
    total_vote_share.values,
    labels=total_vote_share.index,
    autopct="%1.1f%%",
    startangle=90
)
axes[1, 0].set_title("Overall Vote Share (20202025)")

# Histogram  Voter Turnout
axes[1, 1].hist(turnout_values, bins=10, edgecolor="black")
axes[1, 1].set_title("Turnout Percentage Distribution")
axes[1, 1].set_xlabel("Turnout %")
axes[1, 1].set_ylabel("Frequency")

# Layout + Save
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("bihar_election_dashboard.png", dpi=300, bbox_inches="tight")
plt.show()
```
